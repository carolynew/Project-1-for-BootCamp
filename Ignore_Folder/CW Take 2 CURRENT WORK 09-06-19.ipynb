```python
import pandas as pd
from config import api_key
import requests
import json
from yelp.client import Client
```

```python
headers = {'Authorization': 'Bearer %s' % api_key}
```

```python
url='https://api.yelp.com/v3/businesses/search'
 
# In the dictionary, term can take values like food, cafes or businesses like McDonalds
params = {'term':'restaurant','location':'Chicago'}

# Making a get request to the API
req=requests.get(url, params=params, headers=headers)
 
# proceed only if the status code is 200
print('The status code is {}'.format(req.status_code))
```

```python
# printing the text from the response 
sample = json.loads(req.text)
sample['businesses']
```

```python

print(json.dumps(sample, indent=4))
```

```python
# city_dataframe = pd.DataFrame(columns=["Alias", "Name", "Categories",)
city_dataframe = []
for i in range(len(sample["businesses"])):
    try:
        alias_data = sample["businesses"][i]["categories"][0]["alias"]
        name_data = sample["businesses"][i]["name"]
        categories_data = sample["businesses"][i]["categories"][0]["alias"]
        business_rating = sample["businesses"][i]["rating"]
        price_data = sample["businesses"][i]["price"]
    except:
       print("info not found. Skipping...")
       pass
    
    city_dataframe.append({"Alias": alias_data,
                           "Name": name_data,
                           "Categories": categories_data,
                           "Rating": business_rating,
                           "Price": price_data
                         })
```
