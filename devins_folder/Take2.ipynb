```python
miami_data.drop(['businesses','region'], inplace=True)
miami_data
```

```python
height = miami_data.values.flatten()
x_range = range(len(miami_dict))
plt.bar(x_range, height, color='r', alpha=0.5, align='center')

plt.title("Top 5 Cuisine Types in Miami, FL")
plt.xlabel("Cuisine Type")
plt.ylabel("Number of Restaurants")
plt.xticks(x_range, miami_dict.keys())
plt.savefig('Miami Top 5')
plt.show()
```

```python
yelp_api = YelpAPI(api_key)
#search yelp library for Inexpensive Restaurants ($)
miami_search_1_results = yelp_api.search_query(term='restaurants', location='miami, fl', price=1)

#search yelp library for Moderate Restaurants ($$)
miami_search_2_results = yelp_api.search_query(term='restaurants', location='miami, fl', price=2)

#search yelp library for Pricey Restaurants ($$$)
miami_search_3_results = yelp_api.search_query(term='restaurants', location='miami, fl', price=3)

#search yelp library for Ultra High-End ($$$$)
miami_search_4_results = yelp_api.search_query(term='restaurants', location='miami, fl', price=4)
```

```python
pprint(miami_search_1_results['total'])
pprint(miami_search_2_results['total'])
pprint(miami_search_3_results['total'])
pprint(miami_search_4_results['total'])
```

```python
price_point_dict= ({"Inexpensive": miami_search_1_results,
                "Moderate": miami_search_2_results,
                "Pricey": miami_search_3_results,
                "High End": miami_search_4_results   
})

price_data = pd.DataFrame(price_point_dict)
price_data
```

```python
# Drop businesses & region from dataframe
price_data.drop(['businesses', 'region'], inplace=True)
price_data
```

```python
# Bar chart for pricing in Miami, FL
price_height = price_data.values.flatten()
price_x_range = range(len(price_point_dict))
plt.bar(price_x_range, price_height, color='r', alpha=0.5, align='center')

plt.title("Pricing in Miami, FL")
plt.xlabel("Price Point")
plt.ylabel("Number of Restaurants")
plt.xticks(price_x_range, price_point_dict.keys())
plt.savefig('Pricing in Miami')
plt.show()
```

```python
labels = ["Inexpensive","Moderate","Pricey","high end"]
colors = ["yellow","green","lightblue","orange"]

# Tell matplotlib to create a pie chart based upon the above data
plt.pie(price_data, labels = labels, colors=colors,autopct="%1.1f%%", shadow=True, startangle=140)
# Create axes which are equal so we have a perfect circle
plt.axis("equal")
# Save an image of our chart and print the final product to the screen
plt.savefig("MiamiPrices.png")
plt.show()
```
